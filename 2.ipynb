```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
adventure_works_df = pd.ExcelFile('D:\\daneshkar\\HW-L03-02-statistics\\Adventure Works 2020.xlsx')
```

```python
sales_df = adventure_works_df.parse('Sales')
```

```python
sales_df['week days'] = pd.to_datetime(sales_df['OrderDate']).dt.day_name()

sales_df[['OrderDate', 'week days']].head(10)
```

```python
weekday = ['Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday']
sales_df['week days'] = pd.Categorical(sales_df['week days'], categories = weekday, ordered=True)
week_days_count = sales_df['week days'].value_counts().reset_index().sort_values(by = 'week days')

week_days_count
```

```python
fig,ax = plt.subplots(figsize=(16,7))
ax.plot(weekday, week_days_count['count'], marker = 'o')
plt.title('Number of Orders by Week Days' , color = 'gray' , fontsize = 20 , loc = 'left')
ax.grid(axis='y')
ax.spines[['top','right']].set_visible(False)
ax.tick_params(axis='x' , labelsize = 'large')
ax.tick_params( axis = 'x' , which = 'major' , length = 0)
plt.show()
```
