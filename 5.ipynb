```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
adventure_works_df = pd.ExcelFile('D:\\daneshkar\\HW-L03-02-statistics\\Adventure Works 2020.xlsx')
```

```python
sales_df = adventure_works_df.parse('Sales')
customers_df = adventure_works_df.parse('Customers')

merged_df = pd.merge(customers_df , sales_df , how = 'left' ,on = 'CustomerKey')

merged_df[[ 'Name', 'CustomerKey' , 'OrderDate' ]].sort_values(by='OrderDate', ascending=False).reset_index().head()
```

```python
last_orders = merged_df.groupby(['Name' , 'CustomerKey'])['OrderDate'].max().reset_index()

last_orders.head()
```

```python
#this is measured relative t otoday

last_orders['OrderDate'] = pd.to_datetime(last_orders['OrderDate'])

last_orders['DormantCustomers'] = (pd.Timestamp.today() - last_orders['OrderDate']) > pd.Timedelta(days = 180)

last_orders =last_orders.loc[last_orders['DormantCustomers'] == True]

last_orders.head()
```

```python
#this is relative to the last order date of the entire store

max_order = last_orders['OrderDate'].max()

last_orders['DormantCustomers'] = (max_order - last_orders['OrderDate']) > pd.Timedelta(days = 180)

last_orders =last_orders.loc[last_orders['DormantCustomers'] == True]

last_orders.head()
```
